```python
from sqlalchemy.engine import URL 
from sqlalchemy import create_engine
from getpass import getpass

url = URL.create(
    drivername="mysql+pymysql",
    username = "root",
    password = getpass("Enter MySQL password: "), # Prompt user for password
    host="localhost",
    port=3306,
    database="job_posting_db"
)

engine = create_engine(url)
```

```python
import pandas as pd

# Excel File Path
file_path = "../Datasets/New Folder/job_postings_fact.xlsx"

# Read workbook
excel_file = pd.ExcelFile(file_path)

# Load each sheet into MySQL
for sheet in excel_file.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet)

    # Clean table name
    table_name = sheet.lower()

    # Write to MySQL
    df.to_sql(
        name=table_name,
        con=engine,
        if_exists="fail",
        index=False
    )

    print(f"Loaded {sheet} into table '{table_name}'")

print("All sheets loaded successfully!")
```
